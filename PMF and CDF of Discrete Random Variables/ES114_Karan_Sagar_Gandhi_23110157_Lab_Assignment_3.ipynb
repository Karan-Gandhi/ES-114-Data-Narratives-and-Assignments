## Visualizing various Probability Mass Functions for different Random Variables

Here, I plot a bar graph for different random random variables. First, I'll define the necessary functions and import the required modules.

```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
# This is our standard factorial function that is used to calculate the factorial of a number
def factorial(n):
  res = 1
  for i in range(1, n + 1):
    res *= i
  return res

# This function will be used to plot a bar graph for the given PMF of the random variable.
# It takes in the PMF x, the title of the graph and the first random variable as an argument.
def PlotRandomVariablePMF(x, title, start=0):
  plt.figure()
  # Provide the labels and the heights of each element of the bar graph
  plt.bar([str(i) for i in np.arange(start, start + len(x))], x)
  # limit the range of values that the y-axis takes to [0, 1]
  plt.ylim(0, 1)
  plt.xlabel("Random Variable")
  plt.ylabel("Probability $p$")
  plt.title(title)
```

### Discrete Uniform Random Variable

This type of PMF is obtained when all the **distinct** random variables have the same probability or are equiprobable. The PMF is defined as:

$$P_x[k] = \frac{1}{n}$$

Where $n$ is the number of random variables.

```python
def DiscreteUniform(n):
  return [1/n for _ in range(n)]
```

```python
# This plots a Discrete Uniform Distribution with n = 5.
n1 = 5
x1 = DiscreteUniform(n1)
PlotRandomVariablePMF(x1, f"Discrete Uniform Random Variable PMF with n = {n1}")
# This plots a Discrete Uniform Distribution with n = 7.
n2 = 7
x2 = DiscreteUniform(n2)
PlotRandomVariablePMF(DiscreteUniform(n2), f"Discrete Uniform Random Variable PMF with n = {n2}")
plt.show()
```

### Bernoulli Random Variable
Bernoulli random variable is used when we have only two random variables, and one of them has a probability of $p$ of occurring. For example, flipping a coin.

$$ P_x[k] =   \left\{
\begin{array}{ll}
      1 - p & \text{if }k = 0 \\
      p & \text{if }k = 1 \\
\end{array}
\right.  $$

```python
def Bernoulli(p):
  return [1 - p, p]
```

```python
# This plots a Bernoulli Distribution with p = 1/7.
p1 = 1/7
x1 = Bernoulli(p1)
PlotRandomVariablePMF(x1, f"Bernoulli Random Variable PMF with p = {p1}")
# This plots a Bernoulli Distribution with p = 1/3.
p2 = 1/3
x2 = Bernoulli(p2)
PlotRandomVariablePMF(x2, f"Bernoulli Random Variable PMF with p = {p2}")
plt.show()
```

### Binomial Random Variable

This is used to determine the probability of getting $k$ successes out of $M$ independent trials.

$$ P_x[k] = {{M}\choose{k}} \ p^k \cdot (1 - p) ^ {M - k} $$

```python
def Binomial(M, p):
  return [factorial(M)/(factorial(k) * factorial(M - k)) * (p ** k) * ((1 - p) ** (M - k)) for k in range(M + 1)]
```

```python
# Plots a binomial distribution with M = 7 and p = 1/4
M1 = 7
p1 = 1/4
x1 = Binomial(M1, p1)
PlotRandomVariablePMF(x1, f"Binomial Random Variable PMF with M = {M1} and p = {p1}")

# Plots a binomial distribution with M = 7 and p = 1/2
M2 = 7
p2 = 1/2
x2 = Binomial(M2, p2)
PlotRandomVariablePMF(x2, f"Binomial Random Variable PMF with M = {M2} and p = {p2}")

plt.show()
```

### Poisson Random Variable
This is used when you want to determine the probability of getting $k$ success out of $M$ independent trials when M tends to infinity and the probability $p$ of a given success tends to 0.

$$P_x[k] = \frac{e ^ {-\lambda} \ \lambda^{k}}{k!}$$

Where $\lambda$ is defined as:

$$λ = p\cdot M$$

```python
# This will return the a poisson distribution for all values of k in [0, k_max].
def Poisson(l, k_max):
  return [math.exp(-l) * (l ** k) / factorial(k) for k in range(k_max + 1)]
```

```python
k_max = 20
# Plot a poisson distribution with lambda = 10 and k_max = 20.
l1 = 10
x1 = Poisson(l1, k_max)
PlotRandomVariablePMF(x1, f"Poisson Random Variable PMF with λ = {l1}")

# Plot a poisson distribution with lambda = 3 and k_max = 20.
l2 = 3
x2 = Poisson(l2, k_max)
PlotRandomVariablePMF(x2, f"Poisson Random Variable PMF with λ = {l2}")

plt.show()
```

### Geometric Random Variable

This is used to find the probability of getting first success after k independent trials.

$$P_x[k] = (1 - p)^{k - 1} p$$

```python
def Geometric(p, k_max):
  return [((1 - p) ** (k - 1)) * p for k in range(1, k_max + 1)]
```

```python
k_max = 15
# Plot a Geometric distribution with p = 1/3 and k_max = 15.
p1 = 1/3
x1 = Geometric(p1, k_max)
PlotRandomVariablePMF(x1, f"Geometric Random Variable PMF with p = {p1}", start=1)

# Plot a Geometric distribution with p = 1/5 and k_max = 15.
p2 = 1/5
x2 = Geometric(p2, k_max)
PlotRandomVariablePMF(x2, f"Geometric Random Variable PMF with p = {p2}", start=1)

plt.show()
```

## Verifying the equivalence of the Binomial Distribution and the Poisson Distribution for large values of $M$ and small values of $p$

Over here, I take $M$ to be $150$ and $p$ to be $0.01$; hence $\lambda = M \cdot p = 1.5$. Plotting the PMF for both types of random variables shows very little difference. The Mean Absolute difference between the two graphs comes out to be around: `3.478e-05`. This shows that for large values of $M$ and small values of $p$, Binomial and Poisson are equivalent.

```python
M = 150
p = 0.01
l = M * p

# Get the pmf
binomial_rv = np.array(Binomial(M, p))
poisson_rv = np.array(Poisson(l, M))

# Make 2 subplots that are side by side.
fig, ax = plt.subplots(1, 2, sharex=False, sharey=False, figsize=(20, 5))

# Plot binomial_rv
ax[0].bar(np.arange(len(binomial_rv)), binomial_rv)
ax[0].set_xlabel("Random Variable")
ax[0].set_ylabel("Probability $p$")
ax[0].set_title(f"Binomial Random Variable PMF with $M = {M}$ and $p = {p}$")

# Plot poisson_rv
ax[1].bar(np.arange(len(poisson_rv)), poisson_rv)
ax[1].set_xlabel("Random Variable")
ax[1].set_ylabel("Probability $p$")
ax[1].set_title(f"Poisson Random Variable PMF with $λ = {l}$")

plt.show()
```

You can clearly see that both the distributions are almost identical when we plot them on top of each other or plot the difference betweeen the PMFs

```python
# Plot both poisson_rv and binomial_rv on top of each other.
plt.figure()
plt.bar(np.arange(len(binomial_rv)), binomial_rv)
plt.bar(np.arange(len(poisson_rv)), poisson_rv)
plt.legend(["Binomial Random Variable", "Poisson Random Variable"])
plt.xlabel("Random Variable")
plt.ylabel("Probability $p$")
plt.title(f"PMF of both the Distributions plotted together (M = {M}, p = {p}, λ = {l})")

# Plotting the difference between the two random variables
plt.figure()
plt.plot(binomial_rv - poisson_rv)
plt.xlabel("Random Variable")
plt.ylabel("Difference in probability $\Delta p$")
plt.title("Difference between the two PMFs")

plt.show()
```

Further, we can prove that the Distributions are almost the same by calculating the Mean absolute difference between the two PMFs.

```python
mad = np.mean(np.abs(binomial_rv - poisson_rv))
print(f"The Mean Absolute difference comes out to be: {mad}")
```
